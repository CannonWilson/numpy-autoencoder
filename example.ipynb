The following experiments will show off the
autoencoder using both relu and sigmoid 
activation functions. Visualizations are
from the first 5 images in the test set.

```python
from autoencoder import AutoEncoder
```

```python
# Settings constant to both activation functions
INPUT_DIM = 28**2
NUM_EPOCHS = 15
BATCH_SIZE = 64
LR = 1e-3 
LATENT_LAYER = 16
```

```python
# Relu Model
ENCODER_LAYERS = [INPUT_DIM, 512, 256, 128, 64]
DECODER_LAYERS = ENCODER_LAYERS[::-1]
ACTIVATION = "relu"

model = AutoEncoder(NUM_EPOCHS, BATCH_SIZE, LR, ENCODER_LAYERS,\
                        DECODER_LAYERS, LATENT_LAYER, ACTIVATION)
losses = model.train()
model.show_loss_curve(losses)
model.visualize()
```

```python
ENCODER_LAYERS = [INPUT_DIM, 512, 256]
DECODER_LAYERS = ENCODER_LAYERS[::-1]
ACTIVATION = "sigmoid"

model = AutoEncoder(NUM_EPOCHS, BATCH_SIZE, LR, ENCODER_LAYERS,\
                        DECODER_LAYERS, LATENT_LAYER, ACTIVATION)
losses = model.train()
model.show_loss_curve(losses)
model.visualize()
```
